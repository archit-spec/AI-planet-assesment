```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel, GPT2Config
from rotary_embedding_torch import RotaryEmbedding
from transformers.models.gpt2.modeling_gpt2 import GPT2Attention, GPT2Model

# Check if a GPU is available and use it
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

class RotaryGPT2Attention(GPT2Attention):
    def __init__(self, config, is_cross_attention=False, layer_idx=None):
        super().__init__(config, is_cross_attention, layer_idx)
        self.rotary_emb = RotaryEmbedding(dim=config.n_embd // config.n_head, use_xpos=True)

    def _attn(self, query, key, value, attention_mask=None, head_mask=None):
        # Apply rotary embeddings with XPos
        query, key = self.rotary_emb.rotate_queries_and_keys(query, key)
        
        # Continue with the original attention mechanism
        attn_weights = torch.matmul(query, key.transpose(-1, -2))
        if attention_mask is not None:
            attn_weights = attn_weights + attention_mask
        attn_weights = torch.nn.functional.softmax(attn_weights, dim=-1)
        if head_mask is not None:
            attn_weights = attn_weights * head_mask
        attn_output = torch.matmul(attn_weights, value)
        return attn_output, attn_weights

class ModifiedGPT2Model(GPT2Model):
    def __init__(self, config):
        super().__init__(config)
        self.wpe = torch.nn.Embedding(config.n_positions, config.hidden_size)
        self.h = torch.nn.ModuleList([RotaryGPT2Attention(config) for _ in range(config.n_layer)])
        self.init_weights()

# Load tokenizer and config
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
config = GPT2Config.from_pretrained("config.json")
config.n_positions = 8192  # Update max_position_embeddings

# Instantiate the modified model
model = GPT2LMHeadModel(config)
model.transformer = ModifiedGPT2Model(config)

# Move the model to GPU
model.to(device)

# Example usage
input_text = "Example text to test the modified model."
input_ids = tokenizer.encode(inp, return_tensors='pt').to(device)

# Generate output
outputs = model.generate(input_ids, max_length=3500)  # Adjust max_length as needed

# Decode generated output
decoded_output = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(decoded_output)
```

```python
from datasets import load_dataset

dataset = load_dataset("akoksal/LongForm",split='train[:10%]')
```

```python
def tokenize_function(examples):
    # Combine input and output into a single string
    prompts = [f"<|endoftext|>Human: {input}\nAssistant: {output}<|endoftext|>" for input, output in zip(examples['input'], examples['output'])]
    
    return tokenizer(prompts, truncation=True, padding=False, max_length=4096)
```

```python
tokenized_datasets = dataset.map(tokenize_function, batched=True, remove_columns=dataset.column_names)
```

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
```

```python
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': tokenizer.eos_token})
```

```python
from transformers import Trainer, TrainingArguments

# Define training arguments
training_args = TrainingArguments(
    output_dir="results",
    num_train_epochs=1,
    per_device_train_batch_size=1,  # Adjust based on your GPU memory
    per_device_eval_batch_size=1,
    evaluation_strategy="steps",
    save_steps=1000,
    warmup_steps=50,
    weight_decay=0.01,
    logging_dir='./logs',
    fp16=False,  # If your GPU supports it
    gradient_accumulation_steps=1,  # Adjust as needed
)

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["input_ids"],
    data_collator=data_collator

)

# Start training
trainer.train()
```
